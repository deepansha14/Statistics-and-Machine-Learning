```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
m = 30
n = 6
```

```python
data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

```python
fig,ax = plt.subplots(1,3,figsize=(15,4))

ax[0].bar(range(n),np.mean(data,axis=0))
ax[0].set_title('Bar plot')

ax[1].errorbar(range(n),np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='s')
ax[1].set_title('Errorbar plot')

ax[2].bar(range(n),np.mean(data,axis=0))
ax[2].errorbar(range(n), np.mean(data, axis=0), np.std(data,axis=0,ddof=1), marker=".",linestyle='',color='b')
ax[2].set_title('Error + bar plot')

plt.show()
```

```python
## Manually specifying the x-axis coordinates

x_axis = [1,2,4,5,6,9]

plt.bar(x_axis, np.mean(data,axis=0))
plt.errorbar(x_axis,np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='.',linestyle='',color='b')
plt.title('Bar+error plot')

plt.show()
```

```python
mat = [[1,3,5,7], [8,3,5,6]]

fig,ax = plt.subplots(nrows=2,ncols=2, figsize=(12,9))

ax[0,0].imshow(mat)

df = pd.DataFrame(mat, columns = ['prop1','prop2','prop3','prop4'])
df.plot(ax=ax[1,0],kind='bar')

ax[0,1].imshow(np.array(mat).T)
df.T.plot(ax=ax[1,1],kind='bar')

plt.show()
```
