```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
n = 1000

# generating a log normal distribution
data = np.exp(np.random.randn(n)/2)

plt.plot(data,'s')
```

```python
k = 15
```

```python
plt.style.use('ggplot')
plt.hist(data, bins=k, alpha=0.5, histtype='stepfilled')
plt.show()
```

```python
# another option with Numpy

y,x = np.histogram(data, bins=k)

xx = (x[1:]+x[:-1])/2

plt.plot(xx,y)
plt.show()
```

```python
# comparing histograms of several distributions:

x1 = np.random.normal(0, 0.8, 100)
x2 = np.random.normal(-2, 1, 100)
x3 = np.random.normal(3, 2, 100)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=12)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```
